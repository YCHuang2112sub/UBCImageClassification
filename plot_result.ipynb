```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
dir_name = 'result'
exp_name = 'efficientnet_b7'
# file_name = 'train_acc_list.pkl'
file_name_list = ['train_acc_list.pkl', 'train_loss_list.pkl', 'valid_acc_list.pkl', 'valid_loss_list.pkl']
```

```python
# print out all the files in the directory
for pth_experiment in Path(dir_name).iterdir():
    print(pth_experiment, ":")
    for pth_setting in pth_experiment.iterdir():
        # check if is folder
        if not pth_setting.is_dir():
            continue
        print("\t", pth_setting)
```

```python
def get_result(path):
    with open(path, 'rb') as f:
        result = pickle.load(f)
    return result
```

```python
def show_acc_loss_result(train_acc_list, val_acc_list, train_loss_list, val_loss_list):
    plt.figure(figsize=(15, 5))
    plt.subplot(121)
    plt.plot(train_acc_list, label='train_acc')
    plt.plot(val_acc_list, label='val_acc')
    plt.legend()
    plt.title('Accuracy')

    plt.subplot(122)
    plt.plot(train_loss_list, label='train_loss')
    plt.plot(val_loss_list, label='val_loss')
    plt.legend()
    plt.title('Loss')
```

```python
train_acc_list = get_result(Path(dir_name, exp_name, 'train_acc_list.pkl'))
train_loss_list = get_result(Path(dir_name, exp_name, 'train_loss_list.pkl'))
val_acc_list = get_result(Path(dir_name, exp_name, 'valid_acc_list.pkl'))
val_loss_list = get_result(Path(dir_name, exp_name, 'valid_loss_list.pkl'))
```

```python
show_acc_loss_result(train_acc_list, val_acc_list, train_loss_list, val_loss_list)
```
